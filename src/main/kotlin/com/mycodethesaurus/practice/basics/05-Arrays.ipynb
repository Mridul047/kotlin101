```kotlin

// 23 is Int

fun findMax(nums: IntArray): Int {
    var maxNum = Int.MIN_VALUE
    for (num in nums) {
        if (num > maxNum) {
            maxNum = num
        }
    }
    return maxNum
}

fun uniqueRandomIntArray(size: Int, range: IntRange): IntArray {
    return range.shuffled().take(size).toIntArray()
}


findMax(uniqueRandomIntArray(40, 1..1000))
```
