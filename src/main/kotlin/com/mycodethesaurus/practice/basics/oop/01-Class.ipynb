```kotlin
// Class with primary constructor and properties
class Product(
    val id: String,
    var name: String,
    var price: Double,
    var stock: Int
) {
    fun isAvailable(): Boolean = stock > 0
}

// Data class is used to as DTOs
data class OrderSummary(
    val orderId: String,
    var amount: Double = 0.0,
    var items: Int = 0
)

class Cart {
    private val items = mutableListOf<Product>()

    fun addItem(product: Product) {
        if (product.isAvailable()) items.add(product)
    }

    fun getItemCount(): Int = items.size

    fun getTotalAmount(): Double =
        items.sumOf { it.price }
}

open class User(
    val id: String,
    val name: String
)

class Customer(id: String, name: String, val address: String) : User(id, name)

class Admin(id: String, name: String, val permissions: List<String>) : User(id, name)


val customer1 = Customer("C1001", "Tim", "Delhi")
var customer1Cart = Cart()
var customer1OrderSummary = OrderSummary("Order-01-C1001")

var laptop = Product("P1001", "MacBook Air", 999.0, 5)


if (laptop.isAvailable()) {
    // Add Item to cart
    customer1Cart.addItem(laptop)

    //Update Order Summary
    customer1OrderSummary.items = customer1Cart.getItemCount()
    customer1OrderSummary.amount = customer1Cart.getTotalAmount()

    // Print the order
    println("Customer: ${customer1.id} \nOrder Summary: ${customer1OrderSummary}")
}
```

```kotlin
import java.time.LocalDate
import java.util.Currency
import java.util.Locale
import java.util.UUID

// User interaction with Bank Account

class User(
    val userId: String,
    val firstName: String,
    val lastName: String,
    val dateOfBirth: LocalDate,
    val emailId: String
) {
}

interface Account {
    var accountId: String
    var currency: Currency
    var branchName: String
    var userId: String
    var accountBalance: Double
    var transactions: MutableList<Double>

    fun printAccountInformation()
    fun deposit(amount: Double)
    fun withdraw(amount: Double)
}

class SavingsAccount(
    override var accountId: String,
    override var currency: Currency,
    override var branchName: String,
    override var userId: String,
    override var accountBalance: Double,
    override var transactions: MutableList<Double>
) : Account {

    override fun printAccountInformation() {
        println(
            """SavingsAccountId: $accountId
SavingsAccountCurrency: $currency
SavingsAccountBranchName: $branchName
SavingsAccountUserId: $userId
SavingsAccountBalance: $accountBalance
SavingsAccountTransactions: $transactions
        """
        )

    }

    override fun deposit(amount: Double) {

        if (amount < 0) throw RuntimeException("Please enter valid amount. Amount can't be negative !")
        transactions.add(amount)
        accountBalance += amount
        println("AccountBalance: $accountBalance")
    }

    override fun withdraw(amount: Double) {

        if (amount > accountBalance) throw RuntimeException("Please enter valid amount. Amount can't be greater than account balance!")
        transactions.add(-amount)
        accountBalance -= amount
        println("AccountBalance: $accountBalance")
    }

}

var user = User(
    UUID.randomUUID().toString(),
    "mridul",
    "choudhary",
    LocalDate.of(1995, 12, 25),
    "test@gmail.com"
)

var savingsBankAccount = SavingsAccount(
    UUID.randomUUID().toString(),
    Currency.getInstance(
        Locale.US
    ),
    "BOAT CLUB", user.userId, 0.0, mutableListOf()
)

println("=== SavingsBankAccount Intial State ===")
savingsBankAccount.printAccountInformation()

savingsBankAccount.deposit(100.0)
savingsBankAccount.deposit(400.0)
savingsBankAccount.deposit(500.0)
savingsBankAccount.withdraw(300.0)
savingsBankAccount.withdraw(500.0)
//savingsBankAccount.withdraw(900.0)

println("=== SavingsBankAccount Final State ===")
savingsBankAccount.printAccountInformation()
```

```kotlin
// A sealed class restricts which classes can inherit from it
// Think of it as: A controlled hierarchy where all subclasses are known at compile time
// More powerful than ENUMS

fun getData(result: Result) {
    when (result) {
        is Result.Progress -> result.showMessage()
        is Result.Success -> result.showMessage()
        is Result.Error.NonRecoverableError -> result.showMessage()
        is Result.Error.RecoverableError -> result.showMessage()
    }
}

sealed class Result(val message: String) {
    fun showMessage() {
        println("Result: $message")
    }

    class Success(message: String) : Result(message)
    sealed class Error(message: String) : Result(message) {
        class RecoverableError(exception: Exception, message: String) : Error(message)
        class NonRecoverableError(exception: Exception, message: String) : Error(message)
    }

    class Progress(message: String) : Result(message)
}

val success = Result.Success("SUCESS")
val progress = Result.Progress("PROGRESS")

getData(success)
getData(progress)
```

```kotlin
// Object Expression also called as anonymous class

interface OnClickListener {
    fun onClick()
}

// Instead of creating different implementation of interface ise Object Expression
class SignUpClickListener() : OnClickListener {
    override fun onClick() {
        println("Login User")
    }
}

class Button(val text: String, val id: Int, val onClickListener: OnClickListener) {
    fun click() {
        onClickListener.onClick()
    }
}

val loginButton = Button("Login", 1232, object : OnClickListener {
    override fun onClick() {
        println("Login User")
    }
})


val signUpButton = Button("Sign Up", 23423, object : OnClickListener {
    override fun onClick() {
        println("Sign Up User")
    }
})

loginButton.click()
signUpButton.click()
```

```kotlin
import kotlin.properties.ReadWriteProperty
import kotlin.reflect.KProperty

// Delegation on class properties
class User {
    var firstName by FormatDelegate()
    var lastName by FormatDelegate()

}

// Delegate class to get & set string in custom format
class FormatDelegate : ReadWriteProperty<Any?, String> {
    private var formattedStrinq: String = ""

    override fun getValue(
        thisRef: Any?,
        property: KProperty<*>
    ): String {
        return formattedStrinq

    }

    override fun setValue(
        thisRef: Any?,
        property: KProperty<*>,
        value: String
    ) {
        formattedStrinq = value.lowercase()
    }
}

val user = User()
user.firstName = "Jon"
user.lastName = "Snow"

println(user.firstName)
println(user.lastName)
```
