```kotlin
import java.time.LocalTime
import java.time.format.DateTimeFormatter

//Time format -> "16:30:45"
val timeFormatter = DateTimeFormatter.ofPattern("HH:mm:ss")

val currentTime = LocalTime.now().format(timeFormatter)
val parsedCurrentTime = LocalTime.parse(currentTime, timeFormatter)

val hour = parsedCurrentTime.hour

val systemPromt = when (hour) {
    in 1..11 -> "Good Morning Ser! The current time is: $currentTime am"
    in 12..15 -> "Good Afternoon Ser! The current time is: $currentTime pm"
    in 16..18 -> "Good Evening Ser! The current time is: $currentTime pm"
    else -> "Good Night Ser! The current time is: $currentTime pm"
}

println(systemPromt)
```
