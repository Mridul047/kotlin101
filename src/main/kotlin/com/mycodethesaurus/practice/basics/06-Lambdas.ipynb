```kotlin
val add: (Int, Int) -> Unit = { a: Int, b: Int -> println("Sum using Lambda: ${a + b}") }
add(2, 3)
```

```kotlin
val multiply: (Int, Int) -> Unit = { a, b -> println("Product using Lambda: ${a * b}") }
multiply(4, 5)
```

```kotlin
fun add(a: Int, b: Int, action: (Int) -> Unit) {
    action(a + b)
}

add(3, 4, { a: Int -> println(a) })
add(3, 7) { a: Int -> println(a) }
```

```kotlin
// Object Expression & can be replaced with Lambda
class Button(val text: String, val id: Int, val onClickListener: OnClickListener)


interface OnClickListener {
    fun onClick()
}


val loginButton1 = Button("Login", 34345, object : OnClickListener {
    override fun onClick() {
        println("Login Button1 Clicked")

    }
})

val singUpButton1 = Button("Sing Up", 2345, object : OnClickListener {
    override fun onClick() {
        println("Sing Up Button1 Clicked")
    }
})

// Test Login Button1 Click
loginButton1.onClickListener.onClick()

class Button2(val text: String, val id: Int, val onClickListener: () -> Unit)

val loginButton2 = Button2("Login", 34346, {
    println("Login Button2 Clicked")
})

val singUpButton2 = Button2("Sing Up", 2346) {
    println("Sing Up Button2 Clicked")
}

// Test Login Button2 Click
loginButton2.onClickListener()

// Test Singup Button2 Click
singUpButton2.onClickListener()
```

```kotlin
// "it" is an implecit variable created with lambda scope when there is only one parameter in the lambda expression

fun squaredNumbers() {
    val numbers = listOf(1, 2, 3, 4, 5)
    val squaredNumbers = numbers.map { it * it }
    println(squaredNumbers)
}

squaredNumbers()
```

```kotlin
import java.util.UUID

// Scope functions

class User {
    var name: String = ""
    var id: String = ""

    @Override
    fun toStrig(): String {
        return "User(name=$name, id=$id)"
    }
}

var user1 = User()

with(user1) {
    this.name = "Jon"
    this.id = UUID.randomUUID().toString()
}

println(user1.toStrig())

var user2 = User().apply {
    name = "Ned"
    id = UUID.randomUUID().toString()
}

println(user2.toStrig())
```
