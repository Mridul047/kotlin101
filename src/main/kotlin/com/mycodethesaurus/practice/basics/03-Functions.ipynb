```kotlin
val inputNums = (1..10).toList().toIntArray();

fun multiplier(inputNums: Array<Int>) {
    val result = inputNums.map { it * 2 }.toList()
    println("Input number is multiplied by 2: \n $result")
}

multiplier(inputNums);
```

```kotlin
// Traditional way of writting a function
fun getMax(a: Int, b: Int): Int {
    return if (a > b) a else b
}

// Short Hand for writting a function performing single task
fun getMin(a: Int, b: Int): Int = if (a < b) a else b
fun doAddition(a: Int, b: Int): Int = a + b
fun doSubtraction(a: Int, b: Int): Int = a - b

fun getCurrentTime(): String =
    LocalDateTime.now().format(DateTimeFormatter.ofPattern("HH:mm:ss"))

println("Get max number : ${getMax(12, 22)}")
println("Get min number : ${getMin(12, 34)}")
println("Add given numbers : ${doAddition(12, 34)}")
println("Subtract given numbers : ${doSubtraction(12, 34)}")
println("Get current time : ${getCurrentTime()}")
```

```kotlin
import java.time.LocalDateTime
import java.time.format.DateTimeFormatter

// Function with default value defined for a parameter
fun sendMessage(
    name: String,
    message: String = "The is the default message",
    timeStamp: String = LocalDateTime.now()
        .format(DateTimeFormatter.ofPattern("HH:mm:ss"))
) {
    println("Hello $name, $message, system time: $timeStamp")
}

sendMessage("Jim")
sendMessage("Jim", "This is a custom message")
```

```kotlin
fun getSum(vararg nums: Int): Int = nums.sum()

println("Sum of given numbers : ${getSum(12, 3, 4, 1, 1)}")

// * operator before an array is used to expands array into individual arguments
println(
    "Sum of given numbers using array expansion: ${
        getSum(
            *(1..10).toList().toIntArray()
        )
    }"
)
```
