```kotlin
// First & Last number in the range are included

for (i in 1..10) {
    if (i % 2 == 0) {
        println("Event number: $i")
    }
}
```

```kotlin
// Last number in the range is excluded
for (i in 1 until 20) {
    if (i % 2 == 0) {
        println("Event number: $i")
    }
}
```

```kotlin
for (i in 10 downTo 1) {
    if (i % 2 == 0) {
        println("Event number: $i")
    }
}
```

```kotlin
for (i in 1 until 20 step 4) {
    println("Number: $i")
}
```
