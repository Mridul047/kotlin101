```kotlin
// Verbose way of creating singleton class in kotlin

class Database private constructor() {
    companion object {
        private var instance: Database? = null
        fun getInstance(): Database? {
            if (instance == null) {
                instance = Database()
            }
            return instance
        }
    }

    private val connections = mutableListOf<String>()

    fun connect(url: String) {
        connections.add(url)
    }

    fun getConnections(): List<String> = connections
}

val instance1 = Database.getInstance()
val instance2 = Database.getInstance()

println("Check if hashcode is same for both instance:  ${instance1.hashCode() == instance2.hashCode()}")

instance1?.connect("jdbc://localhost:443")
instance1?.getConnections()
```

```kotlin
// Idiomatic way of creating singleton class in kotlin

object Database {
    private val connections = mutableListOf<String>()

    fun connect(url: String) {
        connections.add(url)
    }

    fun getConnections(): List<String> = connections
}

val instance1 = Database
val instance2 = Database

println("Check if hashcode is same for both instance:  ${instance1.hashCode() == instance2.hashCode()}")

instance1.connect("jdbc://localhost:443")
instance1.getConnections()
```
